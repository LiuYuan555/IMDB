Finished basic cleaning, now I want to use Naive Bayes

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['review'], df['sentiment'], test_size = 0.2, random_state=0, stratify=df['sentiment'])
```

```python
X_train.shape
X_test.shape
```

```python
model_nb = Pipeline([
    ('count_vect', CountVectorizer(max_features = 5000, stop_words = 'english')),
    ('naive_bayes', MultinomialNB())
])

model_nb.fit(X_train, y_train)
y_pred = model_nb.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
# Load a specific row (e.g., index 0)
index = 99
actual_review = X_test.iloc[index]
actual_sentiment = y_test.iloc[index]
predicted_sentiment = y_pred[index]

print(f"Review: {actual_review}")
print(f"Actual sentiment: {actual_sentiment}")
print(f"Predicted sentiment: {predicted_sentiment}")
print(f"Correct: {actual_sentiment == predicted_sentiment}")
```

```python
# Find indices where prediction is wrong
wrong_indices = [i for i in range(len(y_test)) if y_test.iloc[i] != y_pred[i]]

print(f"Number of wrong predictions: {len(wrong_indices)}")
print(f"Total predictions: {len(y_test)}")
print(f"Error rate: {len(wrong_indices) / len(y_test) * 100:.2f}%")
```

```python
# Show first 5 wrong predictions
for i in wrong_indices[:5]:
    print(f"\nIndex {i}:")
    print(f"Review: {X_test.iloc[i]}")
    print(f"Actual: {y_test.iloc[i]}")
    print(f"Predicted: {y_pred[i]}")
    print("-" * 50)
```
